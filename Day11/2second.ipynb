```python
import pandas as pd

df = pd.read_csv(r"D:\Python Course\Dealing With Pandas\csv\APPL_1Y_His_Prc - Sheet1.csv")

# Replace Comma
df["Open"] = (df["Open"].astype(str).str.replace(",",".").str.strip())
df["High"] = (df["High"].astype(str).str.replace(",",".").str.strip())
df["Low"] = (df["Low"].astype(str).str.replace(",",".").str.strip())
df["Close"] = (df["Close"].astype(str).str.replace(",",".").str.strip())
df["Volume"] = (df["Volume"].astype(str).str.replace(",",".").str.strip())

# Parse to numeric
df["Open"] = pd.to_numeric(df["Open"], errors="coerce")
df["High"] = pd.to_numeric(df["High"], errors="coerce")
df["Low"] = pd.to_numeric(df["Low"], errors="coerce")
df["CLose"] = pd.to_numeric(df["Close"], errors="coerce")
df["Volume"] = pd.to_numeric(df["Volume"], errors="coerce")

df[:5]
```
