```python
import pandas as pd

df = pd.read_csv(r"D:\Python Course\Dealing With Pandas\csv\APPL_1Y_His_Prc.csv", parse_dates=["Date"], index_col="Date")


# Delete row where has NaN column
df = df[~df.index.isna()]

# Replace Comma 
df["Open"] = (df["Open"].astype(str).str.replace(",",".").str.strip())
df["High"] = (df["High"].astype(str).str.replace(",",".").str.strip())
df["Low"] = (df["Low"].astype(str).str.replace(",",".").str.strip())
df["Close"] = (df["Close"].astype(str).str.replace(",",".").str.strip())
df["Volume"] = (df["Volume"].astype(str).str.replace(",",".").str.strip())

# Parse to numeric
df["Open"] = pd.to_numeric(df["Open"])
df["High"] = pd.to_numeric(df["High"])
df["Low"] = pd.to_numeric(df["Low"])
df["Close"] = pd.to_numeric(df["Close"])
df["Volume"] = pd.to_numeric(df["Volume"])
```
