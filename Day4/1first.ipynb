```python
import pandas as pd


#Read CSV 
df = pd.read_csv(r"D:\Python Course\Dealing With Pandas\csv\messy.csv", na_values={'Age':["-1", "nineteen"]})
df.dropna(axis=1, how="all", inplace=True)
# df.set_index("Name", inplace=True)
df
```

```python
# FillNa
df.fillna({
    'Height(cm)':"NoData",
    'Age':"NoData",
    'Major':"NoData",
    'JoinDate':"NoData"
})

new_df = df.bfill()
new_df["Age"] = new_df["Age"].ffill()
new_df
```

```python
# Fix Interprated
new_df["Name"] = new_df["Name"].str.capitalize()
CleanDf = new_df.set_index("Name")
CleanDf.loc['Nabil']
```
