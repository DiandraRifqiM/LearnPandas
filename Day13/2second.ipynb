```python
import pandas as pd

df = pd.read_csv(r"D:\Python Course\Dealing With Pandas\csv\Day12 - Sheet2.csv", parse_dates=["Date"], index_col=["Date"])
df["Open"] = pd.to_numeric(df["Open"].astype(str).str.replace(",",".").str.strip())
df.head(5)
```

```python
# Shif
df.shift(5)
```

```python
df.shift(-2)
```

```python
df["Prev Per Day"] = df["Open"].shift(1)
df
```

```python
df["1 Day Chnage"] = df["Open"] - df["Prev Per Day"]
df
```

```python
df["5 Day % Return"] = (df["Open"] - df["Open"].shift(5))*100/df["Open"].shift(5)
df
```
