```python
import pandas as pd
dfA = pd.DataFrame({
    'Kota':["Jakarta", "Palembang", "Bogor", "Yogyakarta", "Bandung"],
    'Temp':[29, 27, 26, 30, 25]
})

dfB = pd.DataFrame({
    'Kota':["Jakarta", "Palembang", "Bogor", "Surabaya"],
    'Cuaca':["Cerah", "Berawan", "Hujan Badai", "Cerah"]
})
```

```python
# Inner Join
df1 = pd.merge(dfA, dfB, on="Kota")
df1
```

```python
# Outter Join
df2 = pd.merge(dfA, dfB, how="outer", on="Kota")
df2
```

```python
# Right Join
df3 = pd.merge(dfA, dfB, how="right", on="Kota")
df3
```

```python
# Left Join
df4 = pd.merge(dfA, dfB, how="left", on="Kota")
df4
```

```python
dfC = pd.DataFrame({
    'Kota':["Jakarta", "Palembang", "Bogor", "Yogyakarta", "Bandung"],
    'Temp':[29, 27, 26, 30, 25],
    'Cuaca':["Cerah", "Berawan", "Hujan Badai", "Cerah", "Berawan"]
})

dfD = pd.DataFrame({
    'Kota':["Jakarta", "Bogor", "Surabaya"],
    'Temp':[29, 26, 32],
    'Cuaca':["Cerah", "Berawan", "Cerah"]
})

```

```python
# Suffixes()

df5 = pd.merge(dfC, dfD, on="Kota", suffixes=("Left", "Right"))
df5
```
