```python
import pandas as pd

df = pd.read_csv(r"D:\Python Course\Dealing With Pandas\csv\messy3.csv", encoding="utf-8", parse_dates=["day"])
df["event"] = df["event"].str.capitalize()
df["day"] = df["day"].dt.strftime(r"%d/%m/%Y")
df.columns = df.columns.str.capitalize()
df
```

```python
# Group By Func
dfGroup = df.groupby("City")

# Get Group Data
for city, city_df in dfGroup:
    print(city)
    print(f"{city_df}\n")
```
