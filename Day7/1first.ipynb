```python
import pandas as pd

# Indonesia Weather
indonesia_weather = pd.DataFrame({
    'City':["Jakarta", "Palembang", "Kalimantan"],
    'Temp':[28, 27, 28],
    'Humid':[71, 67, 70 ]
})

indonesia_weather
```

```python
# Japan Weather
japan_weather = pd.DataFrame({
    'City':["Tokyo", "Kyoto", "Shibuya"],
    'Temp':[15, 18, 17],
    'Humid':[46, 52, 48]
})

japan_weather
```

```python

# Concatenate (concat func)
dfConcat = pd.concat([japan_weather, indonesia_weather], ignore_index=True)
dfConcat
```

```python
dfConcat2 = pd.concat([japan_weather, indonesia_weather], keys=["Japan", "Indonesia"])
dfConcat2
```

```python

dfConcat2.loc["Japan"]
```

```python

# Jet Type

type_df = pd.DataFrame({
    'Type':["F22-Raptor", "F16-Falcon", "F15-Eagle"],
    'Brand':["Lockhead Martin", "General Dynamics", "Boeing"]
}, index=[1,2,3])

type_df
```

```python
# Speed

speed_df = pd.DataFrame({
    'Type':["F22-Raptor", "F16-Falcon", "F15-Eagle"],
    'Speed(Mach)':[2, 2.5, 2.5]
}, index=[1,2,3])

speed_df
```

```python

# Concat3
dfConcat3 = pd.concat([type_df, speed_df], axis=1)
dfConcat3
```

```python
lvl = pd.Series(["Beast", "Common", "Rare"], name="Level", index=[1,2,3])
dfConcat4 = pd.concat([dfConcat3, lvl], axis=1)
dfConcat4
```
