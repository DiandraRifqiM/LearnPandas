```python
# Excel
# Change contain "belum | tidak" to "Yet"

def yeet(cell):
    if "belum" in cell.lower():
        return "yeet"
    elif "tidak" in cell.lower():
        return "yeet"
    else:
        return cell
    
df1 = pd.read_excel(r"D:\Python Course\Dealing With Pandas\csv\Questioner IOT.xlsx", header = 3, converters={
    'Question1':yeet,
    'Question2':yeet,
    'Question3':yeet,
    'Question4':yeet,
    'Question5':yeet,
    'Question6':yeet,
    'Question7':yeet,
    'Question8':yeet,
    'Question9':yeet,
    'Question10':yeet
})
df1 = df1.dropna(axis=1, how="all")
df1

```
